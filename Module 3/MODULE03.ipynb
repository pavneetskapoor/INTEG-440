## Flow Control <a id='flow'></a>

This subsection covers [Chapter 2: Flow Control](https://automatetheboringstuff.com/chapter2/) from *Automate the Boring Stuff*. If you are new to Python, be sure to read this chapter first. 

We have already seen how to tell our computer to execute individual instructions, such as evaluate the expression `2 + 2`. Most of the time, however, we don't want our computer to simply execute a series of individual instructions one after the other. Instead, we want to be able to tell our computer to execute instructions *depending on some condition*. This is 'flow control'.

Flow control statements usually include a 'condition' and a 'clause', contained within a 'block'. In the cell below, write a short definition of each. 


#### Condition
Condition is a crucial component that allows for the progression or deterrance of flow of code. Conditions are essentially Boolean values that dictate where the flow of code is to go based on True or False. 

#### Clause 
The block of code that follows a condition is called the clause. An example is the block of code that runs (indented) when a condition of an if statement turns true. 

#### Block
Block is a bundle of lines of codes that follow three characteristics: 1) blocks of code start from indentation 2) a block can contain another block 3) blocks end when indentation returns back to zero 

### <font color="green">YOUR TURN! (Question 3)</font> <a id='yt3'></a>

The cell below executes a simple `if` control flow statement. First, it prompts you to enter the name of this course into a box. Then it executes the statement. In the cell below the code block, translate this control flow statement into plain English. 

```python
print('Please type the name of this course into the box below. ')
course = input()

if course == 'Doing Computational Social Science':
    print('Correct!')
elif course == 'INTEG 440':
    print('Correct!')
elif course == 'INTEG 640':
    print('Correct!')
else:
    print('Incorrect!')
```

The program starts off by printing a string prompting for input through the call 'input()' for typing the name of the course. After, the input that is recorded will be assigned to the variable 'course' which flows into a set of conditional statements that check for the correct course. In the first condition, if the string assigned to 'course' does equal 'Doing Computational Social Science', then the block of conditions will break and the print statement clause will execute to indicate that the name is 'Correct!'. If this is not true, then the next conditional statement is evaluated to see if 'course' is equal to 'INTEG 440' and subsequently for 'INTEG 640'. If all the above conditions are evaluated to be false, then the final clause to be executed will print the statement 'false' indicating the wrong name of the course was provided. 

### <font color="green">YOUR TURN! (Question 4)</font> <a id='yt4'></a>

Question is Worth: <font color="green">.5 points</font>

The cell below executes a simple `while` control flow statement. In the cell below the code block, translate this control flow statement into plain English. 

```python
week = 3

while week < 13:
    print("It's Week {}. The course is still in progress.".format(week))
    week = week + 1

print('\nThe course is complete. Congratulations!')
```

In this `while` control flow statement, the variable `week` is assigned to the number 3. The code then flows into the `while` loop with a terminating condition essentially stating to execute the clause until `week` is less than the number 13. For each iteration, the block of code will print a string stating the week and an update of the course still being in progress. Once week is greater than 13, the terminating condition evaluates to true and the flow goes out of the `while` loop to the print statement that mentions the course being complete.

## Selecting, Aggregating, Summarizing, and Subsetting <a id='sass'></a>

In `Pandas`, each individual column / variable is called a `Series`. To select a `Series` from our `dataframe`, we simply have to type the `Series` name inside square brackets following the name of the `dataframe` object itself. For example, to get the `Series` containing country names, we would type:

```python
df['country']
```

We can perform simple operations on these individual `Series`, such as counting the number of observations per continent in the dataset. Each observation is 1 year of data for a country. There are 12 years of data for each country in our dataset. 

```python
observations_by_continent = df['continent'].value_counts()
observations_by_continent.sort_values(ascending = False)
```

We can create smaller `dataframes` by passing in a list of the `Series` we want to include, or by subsetting the observations based on some criteria. In the first case, we produce a smaller `dataframe` by selecting specific variables. In the second case we produce a smaller `dataframe` by selecting specific observations. For example, if we wanted to select the variables for country, year, and population, we could do the following: 

```python
cpop = df[['country', 'year', 'pop']]
cpop.sort_values(['country', 'year'])
```

The second way we might want to subset our data is by selecting some subset of observations. For example, if we wanted to pull out a subset of our data for Asian countries only, we could do the following: 

```python
asian_data = df[df['continent'] == 'Asia']
asian_data.sample(10)
```

We can identify the unique countries in this dataset using the `.unique()` method on the `Series` country. 

```python
unique_countries_asia = list(asian_data['country'].unique())
len(unique_countries_asia)
```

`Pandas` makes it easy to perform fundamental data operations, such as grouping by one variable and computing a mean value for another. ## Univariate Distributions <a id='dist'></a>
